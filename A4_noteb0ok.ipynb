```python
dictionary = pd.read_csv("assignment_4_files/english-hindi-dictionary.txt",sep="\|\|\|",header=None)
dictionary.columns = ["english","hindi"]
dictionary
```

```python
dictionary["english"] = dictionary["english"].apply(lambda x: x.strip())
dictionary["hindi"] = dictionary["hindi"].apply(lambda x: str(x).strip())

dictionary[dictionary["english"]=="and"]
```

```python
new_biling_dict = {'english':[],'hindi':[],'polarity':[]}
english_list = list(dictionary['english'].values)
hindi_list = list(dictionary['hindi'].values)
for index,row in data.iterrows():
    if row['word'] in english_list and row['word'] not in new_biling_dict['english'] and detect(hindi_list[english_list.index(row['word'])]) == 'hi':
        new_biling_dict['english'].append(row['word'])
        new_biling_dict['hindi'].append(hindi_list[english_list.index(row['word'])])
        new_biling_dict['polarity'].append(row['polarity'])
        
new_biling_dict
```

```python
from gensim.test.utils import datapath
from gensim import utils
def get_data():
    sentences = []
    data = 'assignment_4_files/english.txt'
    for line in open(data):
        sentences.append(utils.simple_preprocess(line))
    return sentences
```

```python
import gensim.models
sentences = get_data()
```

```python
model = gensim.models.Word2Vec(sentences=sentences,min_count=1,workers=6,window=5,size=700,negative=15,iter=20)
```

```python
def get_hindi_data():
    sentences = []
    data = 'assignment_4_files/hindi.txt'
    for line in open(data):
        sentences.append(line.strip().split(" "))
    return sentences
```
