```python
def get_wordnet_tag(word_tag_tuple):
    if word_tag_tuple[1].startswith('J'):
        return wordnet.ADJ
    elif word_tag_tuple[1].startswith('V'):
        return wordnet.VERB
    elif word_tag_tuple[1].startswith('N'):
        return wordnet.NOUN
    elif word_tag_tuple[1].startswith('R'):
        return wordnet.ADV
```

```python
#lemmatized version
import nltk
from nltk.corpus import wordnet
from gensim.test.utils import datapath
from gensim import utils
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
import re
def get_data():
    sentences = []
    data = 'assignment_4_files/english.txt'
    for line in open(data):
        line = re.sub(r'[^\w\s]', '', line)
        line = re.sub("  "," ",line)
        if line:
            try:
                pos_tagged = nltk.pos_tag(str(line).rstrip().lower().split(" "))
                try:
                    sentences.append([wnl.lemmatize(word[0], get_wordnet_tag(word)) for word in pos_tagged])
                except:
                    sentences.append([wnl.lemmatize(word[0]) for word in pos_tagged])
            except:
                pos_tagged = str(line).rstrip().lower().split(" ")

    return sentences
```

```python
#lemmatized
pos_tagged_lexicon = nltk.pos_tag(list(new_bilingiu["english"].values))
lemmatized_set = []
for word in pos_tagged_lexicon:
    try:
        lemmatized_set.append(wnl.lemmatize(word[0], get_wordnet_tag(word)))
    except:
        lemmatized_set.append(wnl.lemmatize(word[0]))
        continue
new_bilingiu['english'] =list(lemmatized_set)
```

```python
import gensim.models
sentences = get_data()
```

```python
model = gensim.models.Word2Vec(sentences=sentences,min_count=3,workers=6,window=5,size=700,negative=10,iter=30)
```

```python
model = gensim.models.Word2Vec(min_count=3,
                     window=5,
                     size=700,
                     sample=6e-5, 
                     alpha=0.03, 
                     min_alpha=0.0007, 
                     negative=10,
                     workers=4)
model.build_vocab(sentences, progress_per=100)
model.train(sentences, total_examples=model.corpus_count, epochs=30, report_delay=1)
```

```python
from langdetect import detect
import numpy as np
def extend_bingliu_lexicon(dictionary,new_bingliu,model,hini_mod):
    new_biling_dict = {'english': list(new_bingliu["english"].values),
                   'hindi': list(new_bingliu["hindi"].values),
                  'polarity':list(new_bingliu["polarity"].values)}
    english_list = list(dictionary['english'].values)
    hindi_list = list(dictionary['hindi'].values)
    for index,row in new_bingliu.iterrows():
        if row["english"] in model.wv.vocab and row["hindi"] in hini_mod.wv.vocab:
            top_5_similar_english = model.wv.similar_by_word(row["english"],5)
            top_5_hindi = hini_mod.wv.similar_by_word(row["hindi"],5)
            top_5_hindi = list(map(lambda x: x[0],top_5_hindi))
    #         print(top_5_similar_english)
            for word in top_5_similar_english:
                word=word[0]
                if word not in new_biling_dict['english'] and word in english_list:
                    hindi_word = hindi_list[english_list.index(word)]
#                     hindi_word = np.array(hindi_list)[[i for i,x in enumerate(english_list) if x==word]].tolist()[0]
#                     print(row["english"],word,hindi_word,top_5_hindi)
#                     print(hindi_word)
                    if hindi_word in top_5_hindi:
                        print(word,hindi_word)
                        new_biling_dict['english'].append(word)

                        new_biling_dict['hindi'].append(hindi_word)
                        new_biling_dict['polarity'].append(row['polarity'])

    return new_biling_dict
     
#     print(top_5_hindi)
        
    
```
